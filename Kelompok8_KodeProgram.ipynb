```python
import os
import cv2
import shutil
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from sklearn.model_selection import train_test_split

# Direktori data asli
dataset_dir = 'dataset/'

# Direktori untuk data yang diaugmentasi
augmented_data_dir = 'augmented_data/'

# Direktori untuk data pelatihan dan pengujian
training_data_dir = 'training_set/'
testing_data_dir = 'testing_set/'

# Fungsi untuk augmentasi data
def augmentasi_data(source_dir, target_dir, augment_size=5):
    datagen = ImageDataGenerator(
        rotation_range=10,
        width_shift_range=0.1,
        height_shift_range=0.1,
        shear_range=0.1,
        zoom_range=0.1,
        fill_mode='nearest'
    )

    for category in ['berlian', 'hati', 'keriting', 'sekop']:
        os.makedirs(os.path.join(target_dir, category), exist_ok=True)
        image_files = os.listdir(os.path.join(source_dir, category))

        for image_file in image_files:
            img_path = os.path.join(source_dir, category, image_file)
            img = cv2.imread(img_path)
            img = img.reshape((1,) + img.shape)
            save_prefix = os.path.splitext(image_file)[0]  # Nama file tanpa ekstensi
            i = 0
            for batch in datagen.flow(img, batch_size=1,
                                      save_to_dir=os.path.join(target_dir, category),
                                      save_prefix=save_prefix, save_format='png'):
                i += 1
                if i >= augment_size:
                    break

# Fungsi untuk membagi dataset menjadi set pelatihan dan pengujian
def split_dataset(source_dir, train_dir, test_dir, test_size=0.2):
    for category in ['berlian', 'hati', 'keriting', 'sekop']:
        os.makedirs(os.path.join(train_dir, category), exist_ok=True)
        os.makedirs(os.path.join(test_dir, category), exist_ok=True)
        image_files = os.listdir(os.path.join(source_dir, category))
        image_files = [os.path.join(source_dir, category, f) for f in image_files]
        train_files, test_files = train_test_split(image_files, test_size=test_size, random_state=42)
        
        # Salin file ke direktori pelatihan dan pengujian
        for file in train_files:
            shutil.copy(file, os.path.join(train_dir, category))
        for file in test_files:
            shutil.copy(file, os.path.join(test_dir, category))

# Lakukan augmentasi data
augmentasi_data(dataset_dir, augmented_data_dir, augment_size=5)

# Membagi data augmentasi menjadi set pelatihan dan pengujian
split_dataset(augmented_data_dir, training_data_dir, testing_data_dir, test_size=0.2)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, Dropout
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.preprocessing.image import ImageDataGenerator

# Ukuran gambar yang akan digunakan dalam model
image_width, image_height = 64, 64

# Inisialisasi model CNN
model = Sequential()

# Menambahkan lapisan konvolusi dan max pooling
model.add(Conv2D(32, (3, 3), input_shape=(image_width, image_height, 3), activation='relu'))
model.add(MaxPooling2D(pool_size=(2, 2)))

# Menambahkan lapisan konvolusi kedua dan max pooling
model.add(Conv2D(64, (3, 3), activation='relu'))
model.add(MaxPooling2D(pool_size=(2, 2)))

# Menambahkan lapisan konvolusi ketiga dan max pooling
model.add(Conv2D(128, (3, 3), activation='relu'))
model.add(MaxPooling2D(pool_size=(2, 2)))

# Flattening
model.add(Flatten())

# Full connection
model.add(Dense(units=128, activation='relu'))
model.add(Dropout(0.5))
model.add(Dense(units=4, activation='softmax'))  # 4 units untuk 4 jenis kartu

# Kompilasi model
model.compile(optimizer=Adam(learning_rate=0.0001), loss='categorical_crossentropy', metrics=['accuracy'])

# Augmentasi data dan persiapan generator data
train_datagen = ImageDataGenerator(rescale=1./255,
                                   rotation_range=10,
                                   width_shift_range=0.1,
                                   height_shift_range=0.1,
                                   shear_range=0.1,
                                   zoom_range=0.1,
                                   horizontal_flip=True)

test_datagen = ImageDataGenerator(rescale=1./255)

# Data generator untuk set pelatihan
train_generator = train_datagen.flow_from_directory(
    training_data_dir,
    target_size=(image_width, image_height),
    batch_size=32,
    class_mode='categorical')

# Data generator untuk set pengujian
validation_generator = test_datagen.flow_from_directory(
    testing_data_dir,
    target_size=(image_width, image_height),
    batch_size=32,
    class_mode='categorical')

# Melatih model
history = model.fit(
    train_generator,
    steps_per_epoch=train_generator.samples // 32,
    epochs=50,
    validation_data=validation_generator,
    validation_steps=validation_generator.samples // 32)

# Menyimpan model
model.save('playing_cards_cnn_model.h5')

# Evaluasi model
eval_result = model.evaluate(validation_generator)
print(f'Test Loss: {eval_result[0]}, Test Accuracy: {eval_result[1]}')
```

```python
import numpy as np
import matplotlib.pyplot as plt
from tensorflow.keras.applications import VGG16
from tensorflow.keras.layers import Dense, Flatten, Dropout
from tensorflow.keras.models import Sequential
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.preprocessing.image import ImageDataGenerator

# Jalur (Path) untuk data pelatihan (training) dan pengujian (testing)
training_data_dir = 'training_set/'
testing_data_dir = 'testing_set/' 

# ImageDataGenerator untuk augmentasi data
train_datagen = ImageDataGenerator(
    rescale=1./255,
    rotation_range=20,
    width_shift_range=0.2,
    height_shift_range=0.2,
    shear_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True,
    fill_mode='nearest'
)

test_datagen = ImageDataGenerator(rescale=1./255)

# Mengalirkan (Flow) gambar pelatihan dalam batch menggunakan generator train_datagen
train_generator = train_datagen.flow_from_directory(
    training_data_dir,
    target_size=(128, 128),
    batch_size=32,
    class_mode='categorical'
)

# Mengalirkan (Flow) gambar pengujian dalam batch menggunakan generator test_datagen
validation_generator = test_datagen.flow_from_directory(
    testing_data_dir,
    target_size=(128, 128),
    batch_size=32,
    class_mode='categorical'
)

# Memuat model VGG16, yang telah dilatih pada data ImageNet
base_model = VGG16(weights='imagenet', include_top=False, input_shape=(128, 128, 3))

# Membekukan (Freeze) lapisan pada model dasar
for layer in base_model.layers:
    layer.trainable = False

# Membuat modelnya
model = Sequential([
    base_model,
    Flatten(),
    Dense(512, activation='relu'),
    Dropout(0.5),  # Dropout to reduce overfitting
    Dense(4, activation='softmax')  # 4 classes
])

# Kompilasi model
model.compile(optimizer=Adam(learning_rate=0.0001), loss='categorical_crossentropy', metrics=['accuracy'])

# Melatih modelnya
history = model.fit(
    train_generator,
    steps_per_epoch=train_generator.samples // train_generator.batch_size,
    validation_data=validation_generator,
    validation_steps=validation_generator.samples // validation_generator.batch_size,
    epochs=50
)

# Menyimpan Model
model.save('model.h5')

# Menggambarkan (Plot) riwayat pelatihan
def plot_history(history):
    acc = history.history['accuracy']
    val_acc = history.history['val_accuracy']
    loss = history.history['loss']
    val_loss = history.history['val_loss']
    epochs = range(1, len(acc) + 1)

    plt.figure(figsize=(15, 5))
    plt.subplot(1, 2, 1)
    plt.plot(epochs, acc, 'bo', label='Training acc')
    plt.plot(epochs, val_acc, 'b', label='Validation acc')
    plt.title('Training and validation accuracy')
    plt.legend()

    plt.subplot(1, 2, 2)
    plt.plot(epochs, loss, 'bo', label='Training loss')
    plt.plot(epochs, val_loss, 'b', label='Validation loss')
    plt.title('Training and validation loss')
    plt.legend()

    plt.show()

plot_history(history)
```

```python
from sklearn.metrics import classification_report, confusion_matrix
import seaborn as sns

# Memastikan bahwa data validasi tidak diacak
validation_generator.shuffle = False
validation_generator.index_array = None

# Memastikan prediksi dan label sebenarnya
predictions = model.predict(validation_generator, steps=len(validation_generator))
predicted_classes = np.argmax(predictions, axis=1)
true_classes = validation_generator.classes
class_labels = list(validation_generator.class_indices.keys())

# Laporan klasifikasi
report = classification_report(true_classes, predicted_classes, target_names=class_labels)
print(report)

# Confusion matrix
cm = confusion_matrix(true_classes, predicted_classes)
plt.figure(figsize=(10, 10))
sns.heatmap(cm, annot=True, fmt='d', xticklabels=class_labels, yticklabels=class_labels, cmap='viridis')
plt.xlabel('Predicted')
plt.ylabel('True')
plt.show()
```

```python
import cv2
import pytesseract
import numpy as np
from tensorflow.keras.models import load_model
from tensorflow.keras.preprocessing.image import img_to_array
import matplotlib.pyplot as plt

# Melakukan load model klasifikasi suit kartu
suit_model = load_model('model.h5')

# Mendefinisikan nama-nama kelas untuk suit dan karakter
class_names = ['Berlian', 'Hati', 'Keriting', 'Sekop']
character_names = {'A': 'As', '2': 'Dua', '3': 'Tiga', '4': 'Empat', '5': 'Lima', '6': 'Enam', '7': 'Tujuh', '8': 'Delapan', '9': 'Sembilan', '10': 'Sepuluh', 'J': 'Jack', 'Q': 'Queen', 'K': 'King'}

# Fungsi untuk memprediksi suit kartu
def predict_suit(img_path, model):
    img = cv2.imread(img_path)
    img = cv2.resize(img, (128, 128))
    img = img_to_array(img) / 255.0
    img = img.reshape((1,) + img.shape)
    prediction = model.predict(img)
    suit = class_names[np.argmax(prediction)]
    return suit

# Fungsi untuk memperbaiki kesalahan pendeteksian karakter
def correct_ocr_result(detected_text):
    corrections = {
        'S': '5',
        'IO': '10',
        'OQ': 'Q',
        '0': '10',
    }
    return corrections.get(detected_text, detected_text)

# Fungsi untuk memprediksi karakter kartu menggunakan OCR
def predict_character(img_path):
    img = cv2.imread(img_path)
    gray_img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    blur_img = cv2.medianBlur(gray_img, 5)
    thresh_img = cv2.adaptiveThreshold(blur_img, 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C, cv2.THRESH_BINARY_INV, 11, 2)
    cropped_img = thresh_img[:int(thresh_img.shape[0] * 0.19), :int(thresh_img.shape[1] * 0.19)]
    kernel = np.ones((3, 3), np.uint8)
    morph_img = cv2.morphologyEx(cropped_img, cv2.MORPH_CLOSE, kernel)
    custom_config = r'--oem 3 --psm 10 -c tessedit_char_whitelist=0123456789AJQK'
    detected_text = pytesseract.image_to_string(morph_img, config=custom_config).strip().upper()
    corrected_text = correct_ocr_result(detected_text)
    character = character_names.get(corrected_text, "Karakter tidak dikenali")
    return character

# Fungsi utama untuk memprediksi seluruh kartu dan menampilkan gambar dengan subplot
def predict_cards(image_paths, model):
    images = []
    titles = []

    for img_path in image_paths:
        img = cv2.imread(img_path)
        img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)  # Mengubah BGR ke RGB
        images.append(img)

        suit = predict_suit(img_path, model)
        character = predict_character(img_path)
        prediction = f'"{character} {suit}"'
        titles.append(prediction)

    # Menampilkan gambar dengan subplot
    plt.figure(figsize=(15, 10))
    columns = int(np.ceil(len(images) / 2))
    for i in range(len(images)):
        plt.subplot(2, columns, i + 1)
        plt.imshow(images[i])
        plt.title(titles[i])
        plt.axis('off')
    plt.show()

# Path untuk Tesseract OCR 
pytesseract.pytesseract.tesseract_cmd = r'C:\Program Files\Tesseract-OCR\tesseract.exe'

# Gambar yang akan dilakukan pendeteksian pola dan karakter
image_paths = ['data_acak/4D.png', 'data_acak/7S.png', 'data_acak/9H.png', 'data_acak/2C.png', 'data_acak/AH.png', 'data_acak/QS.png']
predict_cards(image_paths, suit_model)
```
